# 20. JSONデータの読み込み

```python
import json
f20 = open('./jawiki-country.json', mode='r')
#d20 = json.load(f20)
d20 = f20.readlines()
f20.close()

# 各オブジェクト（辞書）を1つのリストに格納
ans20 = [json.loads(d) for d in d20]
#for d in d20:
#    ans20.append(json.loads(d))
   
print(ans20[100])
#print(d20)
```

# 21. カテゴリ名を含む行を抽出

```python
import re

pattern = '.*Category.*'
#pattern = '.*日本語国名.*'
for l in ans20:
    lst_data = l["text"].split('\n')
    for d in lst_data:
#        print(d)
        rslt = re.match(pattern, d)
        if rslt: print(d)
```
